The executable file created hull.exe is what you need.

```python
% matplotlib inline
from okean import hull
import numpy as np
import pylab as pl
```

Now replace the hull.hull_exec with the new hull.exe:

```python
hull.hull_exec='/tmp/hull/hull_code/hull.exe'
```

Ok, let us try it:

```python
np.random.seed(0)
x=np.random.rand(100)
y=np.random.rand(100)

j=hull.concave_hull(x,y)
k=hull.convex_hull(x,y)

pl.plot(x,y,'bo')
pl.plot(x[j],y[j],'k')
pl.plot(x[k],y[k],'r');
```

Concave hull can have an extra argument alpha. Let's play with it:'

```python
# first need to get the automatic alpha value:
j,alf=hull.concave_hull(x,y,alpha='ret')

pl.figure(figsize=(8,2.5))
ax=[pl.subplot(1,3,i) for i in range(1,4)]
for i,r in enumerate([1.,.7,2]):
    j=hull.concave_hull(x,y,alpha=alf*r)
    ax[i].plot(x,y,'bo')
    ax[i].plot(x[j],y[j],'r')
```
